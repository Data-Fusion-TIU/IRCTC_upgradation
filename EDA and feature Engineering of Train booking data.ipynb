```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt


```

```python
data = pd.read_csv('train_booking_data.csv')
data
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data['User Age'], kde=True, bins=10)
plt.title('Distribution of User Age')
plt.show()

plt.figure(figsize=(8, 5))
sns.countplot(x='Seat Type', data=data)
plt.title('Seat Type Preferences')
plt.show()

plt.figure(figsize=(8, 5))
sns.countplot(x='Seat Type', hue='Seat Booked (Y/N)', data=data)
plt.title('Booking Status by Seat Type')
plt.show()

plt.figure(figsize=(8, 5))
sns.scatterplot(x='Row', y='Column', hue='Seat Booked (Y/N)', data=data)
plt.title('Seat Location (Row vs Column) and Booking Status')
plt.show()

plt.figure(figsize=(8, 5))
sns.countplot(x='Preferred Age Group', hue='Seat Type', data=data)
plt.title('Age Group vs Seat Type Preferences')
plt.show()

plt.figure(figsize=(10, 6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)

data['Seat Booked (Y/N)'] = data['Seat Booked (Y/N)'].map({'Y': 1, 'N': 0})
data['Preferred Age Group'] = data['Preferred Age Group'].astype('category').cat.codes
data['Seat Type'] = data['Seat Type'].astype('category').cat.codes

```
