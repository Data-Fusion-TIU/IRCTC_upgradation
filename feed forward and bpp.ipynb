```python
import math

def expo(x):
    return 1 / (1 + math.exp(-x))

def expo_dv(x):
    return expo(x) * (1 - expo(x))

x1 = 0.05
x2 = 0.10
w1 = 0.15
w2 = 0.20
w3 = 0.25
w4 = 0.30
w5 = 0.40
w6 = 0.45
w7 = 0.50
w8 = 0.55
b1 = 0.35
b2 = 0.60

do1 = 0.01
do2 = 0.99

# Forward pass
net_h1in = w1 * x1 + w2 * x2 + b1
net_h2in = w3 * x1 + w4 * x2 + b1
print(f" net h1 is { net_h1in} and net h2 is { net_h2in}")
h1out = expo(net_h1in)
h2out = expo(net_h2in)
print(f" h1 is { h1out} and h2 is { h2out}")

net_o1in = w5 * h1out + w6 * h2out + b2
net_o2in = w7 * h1out + w8 * h2out + b2
print(f" net o1 is { net_o1in} and net o2 is { net_o2in}")
o1out = expo(net_o1in)
o2out = expo(net_o2in)
print(f" h1 is { o1out} and h2 is { o2out}")

# Error computation
e1 = 0.5 * (do1 - o1out) ** 2
e2 = 0.5 * (do2 - o2out) ** 2
total_error = e1 + e2

print("Total Error =", total_error)
print()

# Backpropagation
# Compute derivatives of errors with respect to outputs
dE_do1 = o1out - do1
dE_do2 = o2out - do2

# Compute derivatives of outputs with respect to inputs
dO1_dnet_o1in = expo_dv(net_o1in)
dO2_dnet_o2in = expo_dv(net_o2in)

# Compute derivatives of inputs with respect to weights and biases
dnet_o1in_dw5 = h1out
dnet_o1in_dw6 = h2out
dnet_o2in_dw7 = h1out
dnet_o2in_dw8 = h2out
dnet_h1in_dw1 = x1
dnet_h1in_dw2 = x2
dnet_h2in_dw3 = x1
dnet_h2in_dw4 = x2

# Compute derivatives of errors with respect to weights using chain rule
dE_dw5 = dE_do1 * dO1_dnet_o1in * dnet_o1in_dw5
dE_dw6 = dE_do1 * dO1_dnet_o1in * dnet_o1in_dw6
dE_dw7 = dE_do2 * dO2_dnet_o2in * dnet_o2in_dw7
dE_dw8 = dE_do2 * dO2_dnet_o2in * dnet_o2in_dw8
dE_dw1 = (dE_do1 * dO1_dnet_o1in * w5 + dE_do2 * dO2_dnet_o2in * w7) * expo_dv(net_h1in) * dnet_h1in_dw1
dE_dw2 = (dE_do1 * dO1_dnet_o1in * w5 + dE_do2 * dO2_dnet_o2in * w7) * expo_dv(net_h1in) * dnet_h1in_dw2
dE_dw3 = (dE_do1 * dO1_dnet_o1in * w6 + dE_do2 * dO2_dnet_o2in * w8) * expo_dv(net_h2in) * dnet_h2in_dw3
dE_dw4 = (dE_do1 * dO1_dnet_o1in * w6 + dE_do2 * dO2_dnet_o2in * w8) * expo_dv(net_h2in) * dnet_h2in_dw4

# Update weights only
l_rate= 0.5
w5 -= l_rate* dE_dw5
w6 -= l_rate* dE_dw6
w7 -= l_rate* dE_dw7
w8 -= l_rate* dE_dw8
w1 -= l_rate* dE_dw1
w2 -= l_rate* dE_dw2
w3 -= l_rate* dE_dw3
w4 -= l_rate* dE_dw4

print("Updated weights:")
print("w1 =", w1)
print("w2 =", w2)
print("w3 =", w3)
print("w4 =", w4)
print("w5 =", w5)
print("w6 =", w6)
print("w7 =", w7)
print("w8 =", w8)
print("b1 =", b1)  # b1 remains unchanged
print("b2 =", b2)  # b2 remains unchanged
```
